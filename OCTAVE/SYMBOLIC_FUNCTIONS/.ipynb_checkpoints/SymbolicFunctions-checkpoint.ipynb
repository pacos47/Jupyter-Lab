```octave
graphics_toolkit gnuplot
pkg load symbolic
```

```octave
syms x
f = x; % Define the symbolic function
fplot(f); % Plot the function over the default range [-5, 5]
title('Plot of x');
xlabel('x');
ylabel('f(x)');
```

### Customizing the Range
You can specify a custom range for the plot using [xmin xmax]:

```octave
syms x
f = x^2; % Define the symbolic function
fplot(f, [-3, 3]); % Plot over the range [-3 , 3]
title('Plot of x from -10 to 10');
xlabel('x');
ylabel('f(x)');
```

```octave
syms x
f = sin(x) + 2*sin(2*x) + 3*sin(3*x); % Define the symbolic function
fplot(f, [-2*pi, 2*pi]); % Plot over the range [-3 , 3]
title('Plot of x from -10 to 10');
xlabel('x');
ylabel('f(x)');
```
