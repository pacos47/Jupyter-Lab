```octave
% 1. Generate sample points
%syms a b N
%a=-6
%b= 6
N = 20
ts = linspace(-6, 6, N);            % Time samples
f = sin(ts)
% 2. Convert symbolic function to numeric function
%f_numeric = matlabFunction(f);     % Creates a function handle

% 3. Evaluate (sample) the function
%fs = f_numeric(ts);                % Samples of f at ts

% 4. Plot sampled function
plot(ts, f, 'o-');
xlabel('t');
ylabel('f(t)');
title('Sampled Symbolic Function');
grid on;
```

```octave
% Sampling parameters
Fs = 1000;           % Sampling frequency (samples per second)
T = 1/Fs;            % Sampling period
L = 1000;            % Length of signal (number of samples)
x = (0:L-1)*T;       % Time vector

% Define the function
f = sin(x) + 2*cos(2*x) + 3*sin(3*x);

% Compute FFT
Y = fft(f);

% Compute magnitude
P2 = abs(Y/L);       % Two-sided spectrum
P1 = P2(1:L/2+1);    % Single-sided spectrum
P1(2:end-1) = 2*P1(2:end-1); % Scale

% Frequency domain
f_axis = Fs*(0:(L/2))/L;

% Plot
figure;
plot(f_axis, P1)
title('Single-Sided Amplitude Spectrum of f(x)')
xlabel('Frequency (Hz)')
ylabel('|P1(f)|')
grid on;
```

```octave
% Given sample sequence
x = [1 0 -1 0];

% Compute FFT
X = fft(x);

% Compute magnitude
magX = abs(X);

% Frequency vector (for plotting)
N = length(x);
f = (0:N-1)*(1/N); % Normalized frequency

% Plot
stem(f, magX, 'filled')
title('Magnitude Spectrum of x[n] = [1 0 -1 0]')
xlabel('Normalized Frequency')
ylabel('|X[k]|')
grid on
```
