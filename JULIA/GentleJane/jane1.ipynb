#### No need for inheritance, traits, or interfaces.

```julia
### Example: Accepting Anything That's Iterable

function sum_items(iter)
    s = 0
    for x in iter
        s += x
    end
    return s
end
```

```julia

sum_items([1, 2, 3])      # array
```

```julia
sum_items(1:5)            # range
```

```julia
sum_items(Set([1, 2, 3])) # set
### It doesn’t care what the type is — just that it acts like an iterable.
```

#### 📌 Mutating vs Non-Mutating Table
Operation	Non-Mutating	Mutating
Sort	sort(v)	sort!(v)
Reverse	reverse(v)	reverse!(v)
Append/Push	append!(v, [x])	(always mutating)
Replace values	replace(v, ...)	replace!(v, ...)
Broadcasting	y = f.(x)	broadcast!(f, y, x)

```julia
v = [3,1,2]
```

```julia
push(v,5)
```

```julia
push!(v,0)
```

```julia
v
```

```julia
reverse(v)    # without touching it  !!!
```

```julia
v           ### BUT v untouched
```

```julia
reverse!(v)
```

```julia
v
```

```julia
append(v,8)    #exists only as  a append!
```

```julia
append!(v,8)
```

```julia
v
```
